```python
data = pd.read_csv('./data/csv示例文件.csv',encoding='gbk',index_col=0)
data
```

## 2.2 pyMyQSL 读取写入文件

```python
#from sqlalchemy import create_engine
#创建数据库连接
engine = create_engine('mysql+pymysql://root:123456@localhost:3306/day01?charset=utf8') #
#           数据表名     引擎对象  是否把索引写入数据库  如果数据存在,如何处理
#写入数据库 -> my_table -> 数据表名 -> engine -> 引擎对象 -> index=False -> 是否把索引写入数据库 -> if_exists='append' -> 如果数据存在,如何处理
data.to_sql('my_table',engine,index=False,if_exists='append') #append追加 replace 覆盖
print('写入成功！')
```

```python
#读取数据库 -> my_table -> 数据表名 -> engine -> 引擎对象 -> sql_df -> 读取结果
sql_df = pd.read_sql('my_table',engine)
sql_df = pd.read_sql('select name,AKA from my_table limit 0,2',engine)
sql_df
```
