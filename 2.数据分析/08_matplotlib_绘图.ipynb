## 2.1 准备数据并绘制初始折线图

```python
# 1.机器学习概述. 准备x轴(60分钟) 和 y轴(每分钟的温度) 的数据.
x = range(60)                                     # x的值: 0 ~ 60 包左不包右
y_beijing = [random.uniform(15, 18) for i in x]   # i的值: 0 ~ 60 包左不包右

# 2. 创建画布
plt.figure(figsize=(10, 5), dpi=80)

# 3. 开始绘图.
plt.plot(x, y_beijing, color='red')     # color: 颜色

# 4. 具体的绘图动作.
plt.show()
```

## 2.2 添加x轴, y轴刻度信息 以及 网格

```python
# 涉及到的函数:  plt.xticks(x, **kwargs),   plt.yticks(y, **kwargs)
# 1.机器学习概述. 准备x轴(60分钟) 和 y轴(每分钟的温度) 的数据.
x = range(60)                                     # x的值: 0 ~ 60 包左不包右
y_beijing = [random.uniform(15, 18) for i in x]   # i的值: 0 ~ 60 包左不包右

# 2. 创建画布
plt.figure(figsize=(10, 5), dpi=80)

# 3. 开始绘图.
plt.plot(x, y_beijing, color='red')     # color: 颜色

# 4. 添加x轴和y轴的刻度信息.
# 4.1.机器学习概述 准备x轴 刻度信息.
x_ticks_label = [f'11点{i}分' for i in x]        # i的值: 0 ~ 60 包左不包右
y_ticks = range(40)

# 4.2 具体的添加x轴和y轴刻度信息的动作.
plt.xticks(x[::5], x_ticks_label[::5])  # 参1: x轴刻度值   参2: x轴刻度值对应的标签
plt.yticks(y_ticks[::5])                # 参1: y轴刻度值

# 5. 添加网格.
# 参1: 网格线的样式 -    --    -.    :
# 参2: 网格线的透明度
plt.grid(linestyle='--', alpha=0.3)    

# 6. 具体的绘图动作.
plt.show()
```

# 3.在一个画布中多次plot()绘制图形

```python
# 需求: 在一个画布中, 绘制两个城市的 温度图.
# 1.机器学习概述. 准备数据.
x = range(60)
y_beijing = [random.uniform(5, 18) for i in x]
y_shanghai = [random.uniform(10, 15) for i in x]

# 2. 创建画布
plt.figure(figsize=(20, 8), dpi=80)

# 3.具体的绘制动作.
# 3.1.机器学习概述 绘制出两个折线.
plt.plot(x, y_beijing, color='pink', label='北京')
plt.plot(x, y_shanghai, color='blue', label='上海', linestyle='--')

# 3.2 创建x轴的刻度值, 及 x轴, y轴的刻度范围.
x_ticks_label = [f'11点{i}分' for i in x]
y_ticks = range(20)
plt.xticks(x[::5], x_ticks_label[::5])
plt.yticks(y_ticks[::2])

# 3.3 标题, 及 x轴, y轴标签.
plt.title('北京上海1个小时内的气温折线图', fontsize=20)
plt.xlabel('时间', fontsize=12)
plt.ylabel('温度', fontsize=12)

# 3.4 添加网格.
plt.grid(linestyle='--', alpha=0.5)

# 3.5 添加图例.
# plt.legend(loc='best')
plt.legend(loc=0)

# 4. 绘制图形.
plt.show()
```

# 4.一个画布, 多个坐标轴 -> 面向对象绘图方式

```python
# 需求: 在一个画布中, 绘制两个城市的 温度图.
# 1.机器学习概述. 准备数据.
x = range(60)
y_beijing = [random.uniform(5, 18) for i in x]
y_shanghai = [random.uniform(10, 15) for i in x]

# 2. 创建画布
# plt.figure(figsize=(20, 8), dpi=80)
# 以元组的形式返回结果, 即: (画布对象, 坐标轴对象)
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 8), dpi=80)

# 3.具体的绘制动作.
# 3.1.机器学习概述 绘制出两个折线.
# plt.plot(x, y_beijing, color='pink', label='北京')
# plt.plot(x, y_shanghai, color='blue', label='上海', linestyle='--')
axes[0].plot(x, y_beijing, color='pink', label='北京')
axes[1].plot(x, y_shanghai, color='blue', label='上海', linestyle='--')

# 3.2 创建x轴的刻度值, 及 x轴, y轴的刻度范围.
x_ticks_label = [f'11点{i}分' for i in x]
y_ticks = range(20)
# plt.xticks(x[::5], x_ticks_label[::5])
# plt.yticks(y_ticks[::2])

# 设置 左边坐标轴(即: axes[0])
axes[0].set_xticks(x[::5], x_ticks_label[::5])
axes[0].set_yticks(y_ticks[::2])

# 设置 右边坐标轴(即: axes[1.机器学习概述])
# axes[1.机器学习概述].set_xticks(x[::5], x_ticks_label[::5])      # 这个可以拆成两个函数.
axes[1].set_xticks(x[::5])                   # step1: 设置x轴刻度值
axes[1].set_xticklabels(x_ticks_label[::5])  # step2: 设置x轴刻度值对应的标签
axes[1].set_yticks(y_ticks[::2])

# 3.3 标题, 及 x轴, y轴标签.
# plt.title('北京上海1个小时内的气温折线图', fontsize=20)
# plt.xlabel('时间', fontsize=12)
# plt.ylabel('温度', fontsize=12)
# 设置 左边坐标轴(即: axes[0])
axes[0].set_title('北京1个小时内的气温折线图', fontsize=20)
axes[0].set_xlabel('时间', fontsize=12)
axes[0].set_ylabel('温度', fontsize=12)

# 设置 右边坐标轴(即: axes[0])
axes[1].set_title('上海1个小时内的气温折线图', fontsize=20)
axes[1].set_xlabel('时间', fontsize=12)
axes[1].set_ylabel('温度', fontsize=12)

# 3.4 添加网格.
# plt.grid(linestyle='--', alpha=0.5)
axes[0].grid(linestyle='--', alpha=0.5)
axes[1].grid(linestyle='-.', alpha=0.5)

# 3.5 添加图例.
# plt.legend(loc='best')
# plt.legend(loc=0)
axes[0].legend(loc='best')
axes[1].legend(loc='best')

# 3.6 保存图形
plt.savefig('./my_picture2.png')

# 4. 绘制图形.
plt.show()
```

# 5. 简单折线图

```python
# 0.准备数据
x = np.linspace(-10, 10, 1000)  # 等差数列
y = np.sin(x)

# 1.机器学习概述.创建画布
plt.figure(figsize=(20, 8), dpi=100)

# 2.绘制函数图像
plt.plot(x, y)
# 2.1.机器学习概述 添加网格显示
plt.grid()

# 3.显示图像
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


def midpoints(x):
    sl = ()
    for _ in range(x.ndim):
        x = (x[sl + np.index_exp[:-1]] + x[sl + np.index_exp[1:]]) / 2.0
        sl += np.index_exp[:]
    return x

# prepare some coordinates, and attach rgb values to each
r, g, b = np.indices((17, 17, 17)) / 16.0
rc = midpoints(r)
gc = midpoints(g)
bc = midpoints(b)

# define a sphere about [0.5, 0.5, 0.5]
sphere = (rc - 0.5)**2 + (gc - 0.5)**2 + (bc - 0.5)**2 < 0.5**2

# combine the color components
colors = np.zeros(sphere.shape + (3,))
colors[..., 0] = rc
colors[..., 1] = gc
colors[..., 2] = bc

# and plot everything
ax = plt.figure().add_subplot(projection='3d')
ax.voxels(r, g, b, sphere,
          facecolors=colors,
          edgecolors=np.clip(2*colors - 0.5, 0, 1),  # brighter
          linewidth=0.5)
ax.set(xlabel='r', ylabel='g', zlabel='b')
ax.set_aspect('equal')

plt.show()
```

# 6. Matplotlit绘制其它常用图形

## 6.1 柱状图 bar() -> 适用于 对比, 统计...

```python
# 0. 准备数据.
x = ['A', 'B', 'C', 'D']
y = [10, 6, 15, 21]

# 1.机器学习概述. 创建画布.
plt.figure(figsize=(10, 5), dpi=80)

# 2. 绘制柱状图.
# plt.plot(x, y)      # 折线图.
plt.bar(x, y)         # 柱状图

# 2.1.机器学习概述 设置网格
plt.grid(linestyle='--', alpha=0.5)

# 2.2 标题, x轴, y轴标签.
plt.title('柱状图', fontsize=20)
plt.xlabel('分类名', fontsize=12)
plt.ylabel('数量', fontsize=12)

# 3. 显示图像.
plt.show()
```

## 6.2 直方图 hist() -> 适用于 连续的数据的  分布, 统计...

```python
# 0. 准备数据.
data = np.random.randn(500)    # 生成500个随机数 -> 符合正态分布的 

# 1.机器学习概述. 创建画布.
plt.figure(figsize=(10, 5), dpi=80)

# 2. 绘制直方图.
plt.hist(data, bins=5)

# 2.1.机器学习概述 设置网格
plt.grid(linestyle='--', alpha=0.5)

# 2.2 标题, x轴, y轴标签.
plt.title('直方图', fontsize=20)

# 3. 显示图像.
plt.show()
```

## 6.3 饼图 pie() -> 适用于 比例的 统计...

```python
sizes = [25, 35, 25, 15]
labels = ['Category A', 'Category B', 'Category C', 'Category D']

# 1.机器学习概述. 创建画布, 绘制 饼图.
plt.pie(sizes, labels=labels, autopct='%1.机器学习概述.1f%%')  # 绘制饼图，显示百分比
plt.title("Simple Pie Chart")  # 设置图表标题
plt.show()  # 显示图表
```

```python
## 6.4 散点图 scatter() -> 适用于 散点数据的 统计, 或者统计两列值之间的关系....
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.scatter(x, y, color='red')  # 绘制红色散点图
plt.title("Simple Scatter Plot")  # 设置图表标题
plt.xlabel("X-axis")  # 设置X轴标签
plt.ylabel("Y-axis")  # 设置Y轴标签
plt.show()  # 显示图表
```
